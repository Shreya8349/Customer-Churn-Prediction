#### Converting date column to days: 

1. Even though in kaggle challenge it was clearly mentioned that "no subscription within 30days of expiry", considering no subscription within a month of expiry may be more appropriate. This is because membership is mostly monthly basis and not based on 30 days. so some users may skip a month and resubscribe next month. This info will be lost because Train data(january) has 31 days. 30 days or month doesn't matter for april month(kaggle set) 
2. eg: expiry on jan 1 resubscribed on feb 1 but the user has churned according to within 30 days assumption but expiry on april 1 resubscribed on may 1, this user has not churned according to 30 days assumption. This variation between train and test set is not good. Hence I will consider 'no subscription with a month' as a churned user.

#### Transaction_date_moved_up-membership_expire_date:

```python
def fun(x):
    return (((x[1]//10000*365)+(x[1]%10000//100*30.4167)+(x[1]%100))-
            ((x[0]//10000*365)+(x[0]%10000//100*30.4167)+(x[0]%100)))<31
```

### Last_5_not_churns:

```python
#retain only last 5 not churn for each group(ie:msno)
transaction_before_jan_up=transaction_before_jan_up.groupby(['msno']).tail(5)
```

```python
transaction_before_jan_up=transaction_before_jan_up.dropna()
```

```python
#map order of all msno in train to its ascending order position starting from 0
transaction_last_expiry_msno_final=np.load('transaction_last_expiry_msno_final.npy',allow_pickle=True)
transaction_last_expiry_msno_final=transaction_last_expiry_msno_final[:,0]#contains msno in ascending order
```

```python
transaction_last_expiry_msno_final
```

```python
index=np.arange(956327)
```

```python
dic=np.vstack((transaction_last_expiry_msno_final,index)).T
```

```python
dic[0]
```

```python
dic=dict(dic)
```

```python
dic['+++hVY1rZox/33YtvDgmKA2Frg/2qhkz12B9ylCvh8o=']
```

```python
#array to hold last 5 not churn feature
Last_5_not_churns=np.full((956327, 5), -1) #default value -1 meaning no transaction. Also index 5=last1 churn(becasue of asecending order)
count=0
index=-1
for i in range(4239356):
    ind=dic[msno[i]]
    if ind==index:
        count+=1
    else: #arrange values in right position and change 'count' to 0
        if count==3:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][3]
            Last_5_not_churns[index][3]=Last_5_not_churns[index][2]
            Last_5_not_churns[index][2]=Last_5_not_churns[index][1]
            Last_5_not_churns[index][1]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][0]=-1
        elif count==2:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][2]
            Last_5_not_churns[index][3]=Last_5_not_churns[index][1]
            Last_5_not_churns[index][2]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][1]=-1
            Last_5_not_churns[index][0]=-1
        elif count==1:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][1]
            Last_5_not_churns[index][3]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][1]=-1
            Last_5_not_churns[index][0]=-1
        elif count==0:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][0]=-1
        count=0
    index=ind
    Last_5_not_churns[index][count]=Last_not_churn[i]
```
