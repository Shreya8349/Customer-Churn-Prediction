```python
transaction_before_jan['discount']=transaction_before_jan['plan_list_price']-transaction_before_jan['actual_amount_paid']
```

```python
transaction_before_jan[transaction_before_jan['discount']<0]
```

### Observation:
1. Some entries are missing, hence the negative discount.

```python
#Impute -1 for negative values
def fun(x):
    if x<0:
        return -1
    else:
        return x
```

```python
transaction_before_jan['discount']=transaction_before_jan['discount'].apply(fun)
```

```python
transaction_before_jan[transaction_before_jan['discount']<0]
```

```python
transaction_before_jan['per_day_price']=transaction_before_jan['plan_list_price']/transaction_before_jan['payment_plan_days']
```

```python
transaction_before_jan
```

```python
#for missing entries, per day price will  be NaN
#replace it with -1
transaction_before_jan['per_day_price']=transaction_before_jan['per_day_price'].fillna(-1)
```

```python
#https://stackoverflow.com/a/19464054
#check if two consecutive rows of msno and membership column are not same
c=transaction_before_jan[['msno','membership_expire_date']].shift(1) != transaction_before_jan[['msno','membership_expire_date']]
```

```python
c
```

```python
def fun(x):
    return x[0] or x[1] #both false means both msno and membership expiry are same for 2 consecutive rows
```

```python
index=c.apply(fun, axis=1)
```

```python
#delete transactions that didn't have any effect on membership expiry date. 
#Ie. drop 2 consectuives duplicates (membership_expiry column). keep='first'
#(this step can be skipped in this notebook,because it doesn't make difference. 
#but needed for other features like last n churn)
transaction_before_jan=transaction_before_jan[index]
```

```python
transaction_before_jan
```

## Sum_count_and_average_features:

```python
t_sum=transaction_before_jan.groupby(['msno'], as_index=False)['actual_amount_paid','payment_plan_days','discount','per_day_price'].sum()
```

```python
t_sum
```

```python
t_count=transaction_before_jan.groupby(['msno'], as_index=False)['actual_amount_paid','payment_plan_days','discount','per_day_price'].count()
```

```python
t_count
```

```python
t_count[['actual_amount_paid_avg','payment_plan_days_avg','discount_avg','per_day_price_avg']]=t_sum[['actual_amount_paid','payment_plan_days','discount','per_day_price']]/t_count[['actual_amount_paid','payment_plan_days','discount','per_day_price']]
```

```python
t_count=t_count[['msno','actual_amount_paid','actual_amount_paid_avg','payment_plan_days_avg','discount_avg','per_day_price_avg']]
```

```python
t_count.columns=['msno','count','actual_amount_paid_avg','payment_plan_days_avg','discount_avg','per_day_price_avg']
```

```python
t_count
```

```python
#replace negative values by -1 because -1 divided by count gives different values for different users.
def fun(x):
    if x<0:
        return -1
    else:
        return x
```

```python
t_count['discount_avg']=t_count['discount_avg'].apply(fun)
t_count['per_day_price_avg']=t_count['per_day_price_avg'].apply(fun)
```

```python
t_sum=t_sum.merge(t_count,on='msno')
```

```python
t_sum
```

### Observation:
1. Some msnos are missing.

```python
#add missing msnos and order based on train_msno order.
t_sum=pd.DataFrame(Train_msno[:,0],columns=['msno']).merge(t_sum,on='msno',how='left')
```

```python
#replace NaN with -1
t_sum=t_sum.fillna(-1)
```

```python
t_sum=t_sum.drop(['msno'],axis=1)
```

```python
t_sum
```

```python
np.array(t_sum)
```

```python
np.save('transact_sum_count_avg_features',np.array(t_sum))
```

#### Last_5_discount:

```python
#array to hold last 5 not churn feature
Last_5_not_churns=np.full((956327, 5), -1) #default value -1 meaning no transaction. Also index 5=last1 churn(becasue of asecending order)
count=0
index=-1
for i in range(4405223):
    ind=dic[msno[i]]
    if ind==index:
        count+=1
    else: #arrange values in right position and change 'count' to 0
        if count==3:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][3]
            Last_5_not_churns[index][3]=Last_5_not_churns[index][2]
            Last_5_not_churns[index][2]=Last_5_not_churns[index][1]
            Last_5_not_churns[index][1]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][0]=-1
        elif count==2:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][2]
            Last_5_not_churns[index][3]=Last_5_not_churns[index][1]
            Last_5_not_churns[index][2]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][1]=-1
            Last_5_not_churns[index][0]=-1
        elif count==1:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][1]
            Last_5_not_churns[index][3]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][1]=-1
            Last_5_not_churns[index][0]=-1
        elif count==0:
            Last_5_not_churns[index][4]=Last_5_not_churns[index][0]
            Last_5_not_churns[index][0]=-1
        count=0
    index=ind
    Last_5_not_churns[index][count]=Last_5_discount[i]
```
